```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python
index = 200
N = 100
#convergence = 10000
variations = [0, 1, 10, 25, 50, 100, 150, 200]
# 0: 0 A1, all A2
# 1: 1 A1, 199 A2
# 10: 10 A1, 190 A2
# 25: 25 A1, 175 A2
# 50: 50 A1, 150 A2
# 100: 100 A1, 100 A2
# 150: 150 A1, 50 A2
# 200: 200 A1, 0 A2

for trial in range (0,len(variations)):
    conv_100_trials = np.zeros(100)

    initial_input = np.zeros(201)
    n = variations[trial]
    initial_input[n] = 1
    #print(initial_input)
    P = np.zeros((2*N+1, 2*N+1))
    #intermediate = np.zeros(convergence, index + 1); 
    #stores all intermediate state probabilities till convergence
    for i in range(0,index+1):
        for j in range(0, index+1):
            P[i,j] = (comb(2*N,j, exact=True, repetition=False)*(((i)/(2*N))**(j))*
                      ((1-(i)/(2*N))**(2*N-j)) )

    n = 3000 # number of steps to take
    output = np.zeros((n+1,2*N+1)) # clear out any old values
    t = np.arange(0,n)
    output[0,:] = initial_input  #generate first output value 
    for i in range(1,n):
        output[i,:] = np.dot(output[i-1,:],P)
        if np.allclose(output[i,:],output[i-1,:]):



            break  
    
    print('Initial state: A1 = ' + str(variations[trial]) + ',' \
          + ' A2 = ' + str(200 - variations[trial]))
    print('Mean Convergence for 100 trials after '+str(i),'iterations')
    print('Initial state vector x(t):')
    print(output[0,:])
    print('\n')
    print('Steady state vector x(t) after '+str(i)+' steps:  ')
    print(output[i,:])
    print('\n')
```

#### Question 3 Analysis

- The composition of the initial population results in the same steady state vector at the end of the simulation when all of the genes are either A1 or A2.
- However, scenarios when the genes are distributed differently, violate the Perron-Frobenius theorem because the distributions at the convergence do not match in the steady state. 
- If this were to not violate the Perron-Frobenius theorem and the Markov chain ergodic theorem, the steady state would result in a similar vector to the initial state.
