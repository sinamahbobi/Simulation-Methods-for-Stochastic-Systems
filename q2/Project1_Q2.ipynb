```python
import numpy as np
import matplotlib.pyplot as plt

#Simulate tossing a biased coin 200 times where 𝑃[HEAD]= 0.80. Count the number of heads. Record the
# longest run of heads. Generate a histogram for the Bernoulli outcomes.

def flip(p):
    return True if np.random.random() < p else False  

# True = Heads (p > 0.5), False = Tails (p <= 0.5 )

num_heads = 0
num_flips = 200
long_heads = 0
temp_heads = 0
heads_record = []
num_repeats = 10000
# Q1: Simulate tossing a fair coin (a Bernoulli trial) 50 times.  Count the number of heads.  
# Record the longest run of heads.   

for i in range(0,num_flips):
    if flip(0.8):
        num_heads +=1
        temp_heads += 1
        heads_record.append(1)
        if temp_heads >= long_heads:
            long_heads = temp_heads
    else:
        temp_heads = 0
        heads_record.append(0)

print("The number of heads is {}" .format(num_heads))
print ("The longest run of heads {}" .format(long_heads))

f0 = plt.figure(1)
plt.hist(heads_record, bins = 2,  facecolor="orange", edgecolor="black", alpha=0.75)
plt.xlabel('Tails                                             Heads' )
plt.ylabel("Number of Occurrences")
plt.title('Number of Heads and Tails for 200 Flips of Biased Coin')
f0.savefig('BiasedCoin_200.pdf')

heads_arr = []
for i in range(0, num_repeats):
    num_heads = 0
    for j in range(0,num_flips):
        if flip(0.8):
            num_heads +=1
    heads_arr.append(num_heads)


f1 = plt.figure(2) 

bins = np.arange(202)
plt.xlabel("Number of heads")
plt.hist(heads_arr, bins, facecolor="g", edgecolor="black", alpha=0.75)
#plt.xticks(range(203))
plt.xlim([135, 185])
plt.ylabel("Number of Times Value of Heads is Seen")
plt.title("# of Heads that Occurs for Every Trial (200 Flips) for 10000 Trials")
plt.show()
f1.savefig('num_repeats10000.pdf')
```
