### 2. Produce a sequence X by drawing samples from a standard uniform random variable.

##### Compute Cov[ X<sub>k</sub>, X<sub>k+1</sub>]. Are X<sub>k</sub> and X<sub>k+1</sub> uncorrelated? What can you conclude about the independence of X<sub>k</sub> and X<sub>k+1</sub>
               

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0,1, 1000) # 1000 samples of random variable on the interval [0,1]
co_arr = []
x_i = x[0:998]
x_i_1 = x[1:999]
x_cov = np.cov(x_i,x_i_1)
print(x_cov)
plt.scatter(x_i,x_i_1, color = 'orange')
plt.xlabel('$X_{k}$')
plt.ylabel('$X_{k+1}$')
plt.title('$X_{k+1}$ vs. $X_{k}$')
plt.show()
```

- In probability theory and statistics, covariance is a measure of the joint variability of two random variables. Based on this covariance matrix, the covariance of X<sub>k</sub> and X<sub>k+1</sub> is around 0.001. Because it is so close to zero they can be considered to be uncorrelated. 
- In this experiment, every subsequent X<sub>k</sub> is randomly generated independent of the previous X<sub>k</sub>, so X<sub>k</sub> X<sub>k+1</sub> are not only uncorrelated, but are also indpendent. 
- As evident in the plot above, there is no correlation between X<sub>k</sub> and X<sub>k+1</sub>

##### Compute a new sequence Y where Y[k] = X[k] - 2* X[k-1] + 0.5 * X[k-2] - X[k-3]. Assume X[k] = 0 for k ≤ 0. Compute Cov[X<sub>k</sub>, Y<sub>k</sub>]. Are X<sub>k</sub> and Y<sub>k</sub> uncorrelated?

```python
y = []

y.append(x[0])
y.append(x[1]-2*x[0])
y.append(x[2] - 2*x[1] + 0.5 * x[0])

for k in range(3,1000):
    y.append(x[k] - 2*x[k-1] + 0.5* x[k-2] - x[k-3])

x_y_cov = np.cov(x,y)
print(x_y_cov)
plt.scatter(x,y, color = 'red')
plt.xlabel('$X_{k}$')
plt.ylabel('$Y_{k}$')
plt.title('$Y_{k}$ vs. $X_{k}$')
plt.show()
```

- Based on the covariance matrix given above, there is some correlation between X<sub>k</sub> and Y<sub>k</sub>. 
- The covariance of X<sub>k</sub> and Y<sub>k</sub> is around 0.09, which is significant enough to show there is correlation. Since there is correlation, they are not dependence.
- This dependence can also be seen in the plot above, where Y increases slightly as X increases in a linear fashion, with a little bit of noise surrounding the outer edges.
